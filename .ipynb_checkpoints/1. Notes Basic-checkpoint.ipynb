### Basic Graph

- After `x` and `y`, we can do
    - '-'
    - '--'
    - 'o'
    - 'o-'
    - 'o--'
- color
- lw: linewidth
- ms: markersize

````python
x = np.linspace(0, 15, 30)
y = np.sin(x) + 0.1*np.random.randn(len(x))
plt.plot(x, y)
````

````python
plt.plot(x, y, '-')
````

````python
plt.plot(x, y, '--')
````

````python
plt.plot(x, y, 'o-')
````

````python
plt.plot(x, y, 'o--')
````

````python
plt.plot(x, y, 'o')
````

````python
plt.plot(x, y, 'o--', color='blue', lw=0.7)
````

````python
plt.plot(x, y, 'o--', color='blue', lw=0.7, ms=2)
````

Figsize

````python
plt.figure(figsize=(8,3))
plt.plot(x, y, 'o--', color='blue', lw=0.7, ms=2, label='Componet 1')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend(loc='lower right', fontsize=8)
````

Easy to have multiple lines on same plot

````python
x2 = np.linspace(0, 15, 100)
y2 = np.sin(x2)
````

````python
plt.figure(figsize=(8,3))
plt.plot(x, y, 'o--', color='blue', lw=0.7, ms=2, label='Componet 1')
plt.plot(x2, y2, '--', color='green', lw=0.7, ms=2, label='Componet 2')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend(loc='lower right', fontsize=8)
````

Adjust x and y limits

```python
plt.ylim(top=1.5)
```

````python
plt.figure(figsize=(10, 3))
plt.plot(x, y, 'o', color='purple', lw=0.7, ms=5, label='C1')
plt.plot(x2, y2, '-', color='green', lw=0.7, ms=5, label='C2')
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.ylim(top=1.5)
plt.legend(loc='upper right', fontsize=10)
````

adjust the legend
```python
ncol=2
```

````python
plt.figure(figsize=(10, 3))
plt.plot(x, y, 'o', color='purple', lw=0.7, ms=5, label='C1')
plt.plot(x2, y2, '-', color='green', lw=0.7, ms=5, label='C2')
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.ylim(top=1.5)
plt.legend(loc='upper right', fontsize=10, ncol=2)
````
